```python
# questão 2

"""
2. Voltar nas Páginas de um Navegador

Descrição: cada pa gina acessada e empilhada. Ao clicar "voltar", voce retorna a anterior.

Operações:
  • avançar(página): acessar nova pa gina.
  • voltar(): voltar a pa gina anterior.
"""

class Navegador:
    def __init__(self):
        self.historico = []
        self.pagina_atual = "http://roblox.tralalelo.tralala"

    def mostrar_pagina(self):
        print(f"vc estar em {self.pagina_atual}")

    def acessar_pagina(self, pagina):
        self.historico.append(pagina)
        self.pagina_atual = pagina
        print(f"vc acessou {pagina}")

    def voltar(self):
        if len(self.historico) > 0:
            self.pagina_atual = self.historico[-1]
            self.historico.pop()

        else:
            self.pagina_atual = "http://roblox.tralalelo.tralala"
            print("nâo tem nenhuma pagina para voltar")
        self.mostrar_pagina()

# -----------------------------------------------------------------------
# test/demostração do script

RUN = True


navegador = Navegador()
print("navegador iniciado")
navegador.mostrar_pagina()

while RUN:
  print(f"o que deseja fazer? \n1 - acessar pagina \n2 - voltar\n3 - exit")
  opcao = input()
  if opcao == "1":
    pagina = input("qual pagina deseja acessar? ")
    navegador.acessar_pagina(pagina)
  elif opcao == "2":
    navegador.voltar()
  elif opcao == "3":
    print("encerrando navegador")
    RUN = False
  else:
    print("opção invalida")

```

```python
# questão 3

"""
3. Montando um Sanduíche

Descrição: Cada ingrediente e adicionado sobre o anterior. O primeiro a sair sera o u ltimo colocado.
Operações:

  • add_ingrediente(ingrediente): colocar ingrediente.
  • remover_ingrediente(): remover o u ltimo ingrediente colocado.

"""

class Sanduiche:
    def __init__(self):
        print("sanduiche criado")
        self.ingredientes = []

    def adicionar_ingrediente(self, ingrediente):
        print(f"adicionando {ingrediente}")
        self.ingredientes.append(ingrediente)

    def remover_ingrediente(self):
        if len(self.ingredientes) > 0:
            print(f"removendo {self.ingredientes[-1]}")
            self.ingredientes.pop()
        else:
            print("o sanduiche esta vazio")

    def show_sanduiche(self):
        print(f"o sanduiche tem {self.ingredientes}")


# ---------------------------------------------------------------------------------------------
# tests


RUN = True

sanduiche = Sanduiche()

while RUN:
  print(f"o que deseja fazer? \n1 - adicionar ingrediente \n2 - remover ingrediente\n3 - mostra o sanduiche\n4 - exit")
  opcao = input()
  if opcao == "1":
    sanduiche.adicionar_ingrediente(input("ingrediente a ser adicionado: "))
  elif opcao == "2":
    sanduiche.remover_ingrediente()
  elif opcao == "3":
    sanduiche.show_sanduiche()
  elif opcao == "4":
    RUN = False
  else:
    print("opcao invalida")
```

```python

# questão 4
"""
4. Caminho em um Labirinto (Desfazer passos)

Descrição: A cada movimento, o jogador empilha a direça o. Para voltar, vai desempilhando os passos.

Operações:
  • andar(direção): andar.
  • voltar_passo(): voltar um passo.
"""


class Labirinto:
    def __init__(self):
        print("labirinto criado")
        self.caminho = []

    def andar(self, direcao):
        print(f"andando para {direcao}")
        self.caminho.append(direcao)

    def voltar_passo(self):
        if len(self.caminho) > 0:
            print(f"voltando para {self.caminho[-1]}")
            self.caminho.pop()
        else:
            print("o labirinto esta vazio")
    def mostrar_caminho(self):
        if len(self.caminho) > 0:
          print(f"---------- vc ja andou: {self.caminho}  -----------")
        else:
          print("--------- vc nao andou ainda -------")

# -------------------------------------------------------------------------------------------------
#test

RUN = True

labirinto = Labirinto()

while RUN:
    labirinto.mostrar_caminho()
    print(f"o que deseja fazer? \n1 - andar \n2 - voltar passo\n3 - exit")
    opcao = input()
    if opcao == "1":
        labirinto.andar(input("para qual direção? "))
    elif opcao == "2":
        labirinto.voltar_passo()
    elif opcao == "3":
        RUN = False
    else:
        print("opcao invalida")
    print("----------------------------------------------------------\n")
```

```python
# questão 5
"""
5. Pilha de Cartas Viradas

Descrição: Cada carta e colocada sobre a anterior. So e possí vel ver ou remover a do topo.

Operações:
  • add_carta_topo(carta): adicionar carta a pilha.
  • puxar_carta(): virar a carta do topo.
"""

class PilhaDeCartas:
    def __init__(self):
        self.cartas = []

    def add_carta_topo(self, carta):
        print(f"carta adicionada: {carta}")
        self.cartas.append(carta)

    def puxar_carta(self):
        if self.cartas:
            print(f"carta puxada: {self.cartas[-1]}")
            return self.cartas.pop()
        else:
            return None

    def mostrar_cartas(self):
        print(f"cartas: {self.cartas}")

# --------------------------------------
# test

run = True
pilha = PilhaDeCartas()

while run:
  print(f"o que deseja fazer? \n1 - adicionar carta a pilha \n2 - puxar carta\n3 - exit")
  opcao = input()
  pilha.mostrar_cartas()
  if opcao == "1":
    pilha.add_carta_topo(input("qual carta deseja adicionar? "))
  elif opcao == "2":
    pilha.puxar_carta()
  elif opcao == "3":
    run = False
  else:
    print("opcao invalida")
```
